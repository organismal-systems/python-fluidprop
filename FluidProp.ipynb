This worksheet provide a simple way to find the density and viscosity of a fluid at a given temperature.

```python
from fluidprop import FluidProperties
import gsw
import numpy as np
import matplotlib.pyplot as plt # import the pyplot graphics library
plt.ion()                       # use interactive plotting mode
```

```python
temps_water = np.linspace(0.0025,40,50)
temps_seawater = np.linspace(-5,40,50)
temps_air = np.linspace(-30,50,100)
p_atmos = 1.01325 # approximate atmospheric pressure, in bar
```

```python
props_rho0 = [FluidProperties("Water",t,p_atmos).rho[0] for t in temps_water]
#props_seawater = FluidProperties("Water", temps_water,p_atmos)
#props_air = [FluidProperties("Air", temps_air,p_atmos)
gsw_rho35 = [gsw.rho_t_exact(35,t,p_atmos) for t in temps_seawater]
gsw_rho0 = [gsw.rho_t_exact(0,t,p_atmos) for t in temps_water]
```

```python
props_mu0 = [FluidProperties("Water",t,p_atmos).mu[0] for t in temps_water]
gsw_mu35 = [gsw.rho_t_exact(35,t,p_atmos) for t in temps_seawater]
gsw_mu0 = [gsw.rho_t_exact(0,t,p_atmos) for t in temps_water]
```

```python
fig1 = plt.figure()
ax1 = fig1.add_subplot(1,1,1)
ax1.plot(temps_water,props_rho0,label='FP, SA=0')
ax1.plot(temps_water,gsw_rho0,label='GSW, SA=0',linestyle="--")
ax1.plot(temps_water,gsw_rho35,label='GSW, SA=0',linestyle="-")
ax1.grid()
ax1.legend()
```

```python
fig1 = plt.figure()
ax1 = fig1.add_subplot(1,1,1)
ax1.plot(temps_water,props_rho0,label='FP, SA=0')
ax1.plot(temps_water,gsw_rho0,label='GSW, SA=0',linestyle="--")
ax1.plot(temps_water,gsw_rho35,label='GSW, SA=0',linestyle="-")
ax1.grid()
ax1.legend()
```
